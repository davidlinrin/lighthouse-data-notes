```python
tweets = []
for i in range(len(test)):
    doc = json.loads(test[i])
    if ('retweeted_status' in list (doc.keys())):
        if ('extended_tweet' in list (doc['retweeted_status'].keys())):
            tweets.append(doc['retweeted_status']['extended_tweet']['full_text'])
        else:
            tweets.append(doc['text'])
    else:
        tweets.append(doc['text'])
```

```python
len(tweets)
```

```python
data = pd.concat([pd.Series(tweets),pd.Series(np.zeros(len(tweets)))], axis = 1, keys =['text','sentiment'])
```

```python
#data = data.append({'text': ':),asfasfas:(', 'sentiment':0}, ignore_index=True)
```

```python
#(':(' in data['text'][2]) & (':)' in data['text'][2])
```

```python
data['sentiment'] = np.where(data.text.str.contains(":\)") & ~data.text.str.contains(":\("),0,
                            np.where(data.text.str.contains(":\(") & ~data.text.str.contains(":\)"),1, 
                                       np.where(data.text.str.contains(":\(") & data.text.str.contains(":\)"),2,3)))
```

```python
data
```

```python
data = data[data['sentiment']!=3]
data.reset_index(drop=True, inplace = True)
```

```python
data['sentiment'].unique()
```

# Cleaning

```python
features = data['text']
labels = data['sentiment']
```

```python
type(features)
```

```python
len(features)
```

```python
features[0]
```

```python
processed_features = []

for sentence in range(0, len(features)):
    tokens = word_tokenize(features[sentence])
    words = [word for word in tokens if word.isalnum()]
    words = [word for word in words if len(word)>1]
    words = ' '.join(words)
    words = words.lower()
    processed_features.append(words)
```

# Word Embedding

```python
vectorizer = TfidfVectorizer(max_features=2500, stop_words=stopwords.words('english'))
```

```python
X = vectorizer.fit_transform(processed_features)
y = data['sentiment']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

# Model

```python
clf = nb.MultinomialNB()
clf.fit(X_train,y_train)
```

```python
predicted = clf.predict(X_test)
predicted_proba = clf.predict_proba(X_test)[:,1]
predicted
```

```python
confusion_matrix(y_test, predicted)
```

```python
print(f'accuracy: {accuracy_score(y_test,predicted)}')
```

# using NN

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

def plot_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    x = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(x, acc, 'b', label='Training acc')
    plt.plot(x, val_acc, 'r', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(x, loss, 'b', label='Training loss')
    plt.plot(x, val_loss, 'r', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
```

```python
processed_features
```

```python
text_train, text_test, y_train, y_test = train_test_split(processed_features,y)
```

```python
tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(text_train)

X_train = tokenizer.texts_to_sequences(text_train)
X_test = tokenizer.texts_to_sequences(text_test)

vocab_size = len(tokenizer.word_index) + 1  # Adding 1 because of reserved 0 index

print(text_train[2])
print(X_train[2])
```

```python
len(processed_features)
```

```python
vocab_size
```

```python
maxlen = 100

X_train = pad_sequences(X_train, padding='post', maxlen=maxlen)
X_test = pad_sequences(X_test, padding='post', maxlen=maxlen)

print(text_train[0])
print(X_train[0, :])
```

```python
from keras.models import Sequential
from keras import layers

embedding_dim = 50

model = Sequential()
model.add(layers.Embedding(input_dim=vocab_size, 
                           output_dim=embedding_dim, 
                           input_length=maxlen))
model.add(layers.GlobalMaxPool1D())
model.add(layers.Dense(10, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(X_train, y_train,
                    epochs=10,
                    verbose=False,
                    validation_data=(X_test, y_test),
                    batch_size=10)
loss, accuracy = model.evaluate(X_train, y_train, verbose=False)
print("Training Accuracy: {:.4f}".format(accuracy))
loss, accuracy = model.evaluate(X_test, y_test, verbose=False)
print("Testing Accuracy:  {:.4f}".format(accuracy))
plot_history(history)
```

### CNN

```python
embedding_dim = 100

model = Sequential()
model.add(layers.Embedding(vocab_size, embedding_dim, input_length=maxlen))
model.add(layers.Conv1D(128, 5, activation='relu'))
model.add(layers.GlobalMaxPooling1D())
model.add(layers.Dense(10, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(X_train, y_train,
                    epochs=10,
                    verbose=False,
                    validation_data=(X_test, y_test),
                    batch_size=10)
loss, accuracy = model.evaluate(X_train, y_train, verbose=False)
print("Training Accuracy: {:.4f}".format(accuracy))
loss, accuracy = model.evaluate(X_test, y_test, verbose=False)
print("Testing Accuracy:  {:.4f}".format(accuracy))
plot_history(history)
```
