```python
# connect to the database
con = psycopg2.connect(database='drinks', user='postgres', password=postgres_pwd,
                       host='127.0.0.1', port='5432')

query = con.cursor()
```

# Question 2

```python
query.execute(
"""
select bar, avg(price) as avg 
from has_on_menu 
group by bar 
order by avg DESC 
limit 5
""")
    
Q2 = query.fetchall()
```

```python
Q2 = pd.DataFrame(Q2)
Q2.columns =['Bar', 'Average']
Q2
```

# Question 3

```python
query.execute(
"""
select bar, drink_id, price 
from has_on_menu 
order by price
limit 1
""")
    
Q3 = query.fetchall()
```

```python
Q3 = pd.DataFrame(Q3)
Q3.columns =['Bar', 'Drink', ' Price']
Q3
```

# Question 4

```python
query.execute(
"""
select m.bar, Count(*)
from has_on_menu m 
join drink_info d on m.drink_id = d.drink_id
group by m.bar
order by m.bar
""")
    
Q4 = query.fetchall()
```

```python
Q4
```

```python
Q4 = pd.DataFrame(Q4)
Q4.columns =['Bar', 'Drink']
Q4
```

# Challenge Question

```python
For each person, find the bar they visit, and the type(s) and price(s) of the drink(s) they drink during those visits.
```

```python
has_on_menu(bar: string, drink_id: string, price: real)

orders(person: string, date: string, bar, drink_id: string, quantity: integer)

drink_info(drink_id: string, type: string)
```

```python
query.execute(
"""
select o.person, o.bar, d.type, m.price
from orders o
join has_on_menu m on m.drink_id = o.drink_id and m.bar = o.bar
join drink_info d on o.drink_id = d.drink_id
""")
    
Q5 = query.fetchall()
```

```python
Q5 = pd.DataFrame(Q5)
Q5
```
