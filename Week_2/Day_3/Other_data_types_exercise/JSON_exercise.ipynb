```python
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize  
from pprint import pprint
```

```python
file = '/mnt/d/lighthouse/lighthouse_data_notes/Week_2/Day_3/Other_data_types_exercise/world_bank_projects.json'
bank = pd.read_json(file)
```

```python
bank.head()
```

# 1. Find the 10 countries with most projects

```python
projects = bank.groupby('countryname').count()
```

```python
top_project_country = projects.sort_values(by='project_name', ascending = False)
top_project_country['project_name'].head(10)
```

# 2. What are top 10 sectors with projects?# 

```python
bank_sect = bank.explode('sector')
bank_sect.head(6)
```

```python
sector_series = bank_sect['sector']
sector_series
```

```python
sector_values = pd.Series([x.get('Name') for x in sector_series])
sector_values
```

```python
bank_sect['sector'] = sector_values
bank_sect.head(10)
```

```python
top_sectors = bank_sect.groupby('sector').count()
```

```python
top_10_sector = top_sectors.sort_values(by='project_name', ascending = False)
top_10_sector['project_name'].head(10)
```

# 3. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
bank.head(5)
```

```python
bank_theme = bank.explode('mjtheme_namecode')
```

```python
theme_series = bank_theme['mjtheme_namecode']
theme_series
```

```python
theme_values = pd.Series([x.get('code') for x in theme_series])
theme_values.head(13)
```

```python
theme_names = pd.Series([x.get('name') for x in theme_series])
theme_names.head(13)
```

```python
theme = pd.DataFrame()
theme['num'] = theme_values
theme['name'] = theme_names
```

```python
theme
```

```python
theme['name'].replace('', np.nan, inplace=True)
theme.dropna(subset=['name'], inplace=True)
theme_dict = theme.drop_duplicates()
```

```python
theme_dict = theme_dict.set_index('num')['name'].to_dict()
theme_dict
```

```python
theme_values 
```

```python
bank_theme['theme_num'] = theme_values
```

```python
bank_theme['theme_name'] = bank_theme['theme_num'].map(theme_dict)
bank_theme
```

```python
top_theme_num = bank_theme.groupby('theme_name').count()
```

```python
top_10_theme = top_theme_num.sort_values(by='project_name', ascending = False)
top_10_theme['project_name'].head(10)
```
