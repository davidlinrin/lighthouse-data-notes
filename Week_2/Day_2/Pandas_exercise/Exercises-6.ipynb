### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.
- first table is in USD
- second table actually consists of 3 smaller tables. First 9 rows are for USD, next 8 rows are for GBP and final 7 are for EUR.

```python
sale1['currency'] = pd.Series(['USD'] * 5)
sale1
```

```python
USD = pd.Series(['USD'] * 9)
GBP = pd.Series(['GBP'] * 8)
EUR = pd.Series(['EUR'] * 7)
currency = pd.concat([USD,GBP,EUR])
currency = currency.reset_index(drop=True)
```

```python
sale2['Currency'] = currency
sale2_no_na = sale2.dropna()
```

```python
columns = sale1.columns.values
columns
```

```python
sale2.columns=['Book title', 'Number sold', 'Sales price', 'Royalty paid','currency']
```

```python
sale1.head()
```

```python
sale2.head()
```

```python
total = pd.concat([sale1,sale2])
total['Book title'] = total['Book title'].str.lower()
total['Book title'] = total['Book title'].str.replace('[^\w\s]','')
total.columns = total.columns.str.replace(' ', '_')
total.head()
```

```python
US = total[total['currency']=='USD']
US_no_na = US.dropna()
US_no_na.loc[:,'total'] = US_no_na.Number_sold * US_no_na.Sales_price
US_spent = US_no_na.groupby('Book_title').sum()
US_spent
```

```python
GBP = total[total['currency']=='GBP']
GBP_no_na = GBP.dropna()
GBP_no_na.loc[:,'total'] = GBP_no_na.Number_sold * GBP_no_na.Sales_price
GBP_spent = GBP_no_na.groupby('Book_title').sum()
GBP_spent
```

```python
EUR = total[total['currency']=='GBP']
EUR_no_na = EUR.dropna()
EUR_no_na.loc[:,'total'] = EUR_no_na.Number_sold * EUR_no_na.Sales_price
EUR_spent = EUR_no_na.groupby('Book_title').sum()
EUR_spent
```
