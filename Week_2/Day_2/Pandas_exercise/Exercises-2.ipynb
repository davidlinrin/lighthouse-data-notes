### What are the ten most common movie names of all time?

```python
top_10= titles['title'].value_counts()
top_10[:11]
```

### Which three years of the 1930s saw the most films released?

```python
three_year = titles[(titles['year'] <1940) & (1929< titles['year'])]['year'].value_counts()
three_year[:3]
```

### Count the number of films that have been released each decade over the history of cinema.

```python
titles['decade']= (titles['year']//10)*10
decade = titles.groupby('decade').count()
decade
```

### Count the number of "Hamlet" films made each decade.

```python
titles['decade']= (titles['year']//10)*10
hamlet = titles[titles['title'].str.match('Hamlet')]
hamlet_decade = hamlet.groupby('decade').count()
hamlet_decade['title']
```
