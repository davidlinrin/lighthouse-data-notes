### Define a year as a "Superman year" whose films feature more Superman characters than Batman. How many years in film history have been Superman years?

```python
superman = cast[cast['character']=='Superman']
batman = cast[cast['character']=='Batman']
```

```python
superman = superman.groupby('year').count()
superman=superman.rename(columns = {'title':'superman'})
superman.head()
```

```python
batman = batman.groupby('year').count()
batman=batman.rename(columns = {'title':'Batman'})
batman.head()
```

```python
years = pd.concat([superman['superman'],batman['Batman']],axis = 1)
years.head()
```

```python
superman_year = years[years['superman']>years['Batman']]
superman_year
```

### How many years have been "Batman years", with more Batman characters than Superman characters?

```python
batman_year = years[years['superman']<years['Batman']]
batman_year
```

### Count the number of actor roles each year and the number of actress roles each year over the history of film.

```python
roles = cast.groupby(['type','year']).count()
actor = roles.loc['actor']
actress = roles.loc['actress']
actor=actor.rename(columns = {'title':'actor'})
actress=actress.rename(columns = {'title':'actress'})
actor.head()
```

### Find the difference between the number of actor roles each year and the number of actress roles each year over the history of film.

```python
compare = pd.concat([actor['actor'],actress['actress']],axis = 1)
compare = compare.fillna(0)
```

```python
compare['differece'] = abs(compare['actor'] - compare['actress'])
```

```python
compare
```

### What is the fraction of roles that have been 'actor' roles each year in the history of film.

```python
compare['fraction'] = compare['actor']/(compare['actor'] + compare['actress'])
```

```python
compare
```

### What is the fraction of supporting (n=2) roles that have been 'actor' roles each year in the history of film.

```python
support = cast[(cast['n']==2) & (cast['type']=='actor')]
support_year = support.groupby('year').count()
support_year = support_year.rename(columns = {'title':'support'})
support_year.head()
```

```python
main = cast.groupby('year').count()
main = main.rename(columns = {'title':'total'})
main.head()
```

```python
role = pd.concat([support_year['support'],main['total']],axis = 1)
role= role.fillna(0)
role['fraction'] = role['support']/role['total']
```
