```python
file_release = '/mnt/d/lighthouse/lighthouse_data_notes/Week_2/Day_2/Pandas_exercise/release_dates.csv'
release = pd.read_csv(file_release)
release.head()
```

### Count the number of months in which movies with "Christmas" in their title tend to be released in the USA.

```python
release['month'] =  pd.DatetimeIndex(release['date']).month
release['day'] =  pd.DatetimeIndex(release['date']).day
```

```python
christmas_USA = release[(release['title'].str.contains('Christmas')) & (release['country']=='USA')]
christmas_USA
```

```python
USA = christmas_USA .groupby('month').count()
```

```python
USA['title']
```

### Count the number of months in which movies with "Christmas" in their title tend to be released in Canada.
