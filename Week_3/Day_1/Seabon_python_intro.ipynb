```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Regression plot

```python
sns.lmplot(x='total_bill', y='tip', data = tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip', data = tips, hue = 'sex')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip', data = tips, col = 'day', hue ='day')
plt.show()
```

# REsidual plot

```python
sns.residplot(x='total_bill',y='tip', data = tips)
plt.show()
```

# Strip plot

```python
sns.stripplot(x='day',y ='tip',data = tips, size =5, jitter = True, hue = 'sex')

# overwrite the original y-label from 'tip' to 'tip ($)'
plt.ylabel('tip ($)')
plt.show()
```

# Swarm plot

```python
sns.swarmplot(x='day',y='tip',data = tips, hue = 'sex')

# overwrite the original y-label from 'tip' to 'tip ($)'
plt.ylabel('tip ($)')
plt.show()
```

# Box / violin plot

```python
# create a grid with 1 row 2 columns
plt.subplot(1,2,1)

# the first plot
sns.boxplot(x='day',y='tip', data = tips)
plt.ylabel('tip ($)')


plt.subplot(1,2,2)

# the second plot
sns.violinplot(x='day',y='tip', data = tips)
plt.ylabel('tip ($)')
plt.show()
```

# join plot

```python
sns.jointplot(x='total_bill', y='tip', data = tips)
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data = tips, kind = 'kde')
plt.show()
```

# pair plot (scatter plot matrix)

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue = 'sex')
plt.show()
```

# Heat maps

```python
# compute correlations between features
df_corr = tips.corr()

# plot the correlations
sns.heatmap(df_corr)
plt.title('Correlation plot')
plt.show()
```
