```python
# import pandas
import pandas as pd
import numpy as np
# import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv')
```

```python
df.head()
```

# Objective: Read Total profit of all months and show it using a line plot

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.plot('month_number','total_profit',data= df,color='red')
ax.set_title('total profit by month')
ax.set_xlabel('months')
ax.set_ylabel('total profit')
```

# Objective: Read all product sales data and show it  using a multiline plot

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.set_title('total sales by month')
ax.set_xlabel('months')
ax.set_ylabel('sales')
for i in df.columns:
    if i == 'month_number' or i == 'total_profit':
        next
    else:
        ax.plot('month_number', i, data=df, label =i)
plt.legend(loc='upper right')
```

# Objective: Read the total profit of each month and show it using the histogram to see most common profit ranges

```python
fig, ax = plt.subplots(figsize =(10,8))
ax.hist('total_profit',data = df, bins = 7)
ax.set_title('total profit')
ax.set_xlabel('profit')
ax.set_ylabel('freq')
```
