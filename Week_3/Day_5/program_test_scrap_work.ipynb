```python
def compute_statistics(df):
    name = df.groupby('Name').count()
    m = max(name['Age'])
    n = name[name['Age'] == m].index[0]
    m = np.mean(df['Age'])
    m = m.item()
    q75, q25 = np.percentile(df['Rating'], [75 ,25])
    iqr = q75 - q25
    iqr = iqr.item()
    return((n,m,iqr))
```
