# Eigenvectors

```python
A = np.array([[5, 1], [3, 3]])
A
```

```python
# eigen value: 6 with eigenvector [1,2]
v1 = np.array([[1], [1]])
v1
```

```python
Av1 = A.dot(v1) 
Av1
```

```python
np.linalg.eig(A)
```

```python
orange = '#FF9A13'
blue = '#1190FF'

plotVectors([Av1, v1], colors=[blue, orange])
plt.ylim(-1, 7)
plt.xlim(-1, 7)
plt.show()
```

```python
# eigen value: 2 with eigenvector [1,-3]
v2 = np.array([[1], [-3]])
v2
```

```python
A
```

```python
Av2 = A.dot(v2)
Av2
```

```python
plotVectors([Av2, v2], colors=[blue, orange])
plt.ylim(-7, 1)
plt.xlim(-1, 3)
plt.show()
```

```python
eigenvals, V = np.linalg.eig(A)
eigenvals
```

```python
V
```

```python
lambdas = np.diag(eigenvals)
lambdas
```

```python
V_inv = np.linalg.inv(V)
V_inv
```

```python
# spectral decomp, A = PDP^-1 where P is matric of eigenvectors and D is an identiy matrix with eigen values along the diagonal
A_orig  = V.dot(lambdas).dot(V_inv)
A_orig
```
