```python
from pandas import read_csv
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
series = read_csv('airline_passengers.csv', header=0, index_col=0)
series.Passengers.values
```

```python
result = seasonal_decompose(series.Passengers.values, model='multiplicative', freq = 30)
result.plot()
pyplot.show()
```
