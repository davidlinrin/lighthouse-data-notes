# Data

```python
# 9994 rows × 20 columns
dataset_train = pd.read_excel('Sample_Superstore.xls')
dataset_train.drop('Row ID', axis = 1, inplace = True)
dataset_train.head()
```

```python
# 9994 rows × 2 columns
var_interest = 'Profit'
df = dataset_train[['Order Date',var_interest]]
df = df.rename(columns={'Order Date':'ds', var_interest:'y'})
df = df.sort_values('ds')
df.head()
```

```python
X = index_df.values
window = 3
history = [X[i] for i in range(window)]
test = [X[i] for i in range(window, len(X))]
predictions = list()
# walk forward over time steps in test
for t in range(len(test)):
    length = len(history)
    yhat = np.mean([history[i] for i in range(length-window,length)])
    obs = test[t]
    predictions.append(yhat)
    history.append(obs)
    #print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
```

```python
zoom_start = 800
zoom_end = 1000
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(test[zoom_start:zoom_end])
ax.plot(predictions[zoom_start:zoom_end], color='red')
```

```python
X = index_df['y'].values
size = int(len(X) * 0.7)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(0,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    #print('predicted=%f, expected=%f' % (yhat, obs), '\n')
error = mean_squared_error(test, predictions)
```

```python
print('Test MSE: %.3f' % error)
zoom_start = 100
zoom_end = 200
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(test[zoom_start:zoom_end])
ax.plot(predictions[zoom_start:zoom_end], color='red')
```

```python
# 1237 rows × 1 columns
LSTM_df = copy.deepcopy(index_df)
LSTM_df.head()
```

```python
LSTM_df['y-1'] = LSTM_df['y'].shift(1)
LSTM_df['y-2'] = LSTM_df['y'].shift(2)
LSTM_df['y-3'] = LSTM_df['y'].shift(3)
LSTM_df['y-4'] = LSTM_df['y'].shift(4)
LSTM_df['y-5'] = LSTM_df['y'].shift(5)
LSTM_df['y-6'] = LSTM_df['y'].shift(6)
LSTM_df['y-7'] = LSTM_df['y'].shift(7)
```

```python
model = Sequential()

model.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))

model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))

model.add(LSTM(units = 50))
model.add(Dropout(0.2))

model.add(Dense(units = 1))

model.compile(optimizer = 'adam', loss = 'mean_squared_error')

model.fit(X_train, y_train, epochs = 100, batch_size = 32)
```
